```python
from collections import deque       
import copy

def solution(board):
    answer = 0
    n = len(board)
    move_sets = ["up", "down", "right", "left"]
    stack = deque([])
    stack.append(((0,0), "down", [], "", 0))
    stack.append(((0,0), "right", [], "", 0))
    # (start, move_sets, visited, prev, cost)
    
    while stack:
        temp = stack.pop()
        coords, move, visited, prev, cost = temp[0], temp[1], temp[2], temp[3], temp[4]
        next_visited = copy.deepcopy(visited)
        next_visited.append(coords)
        
        if move == move_sets[0]: #up
            nc = (coords[0] - 1, coords[1])
            if prev == "hor":
                cost = cost + 600
            else:
                cost = cost + 100
            next_prev = "ver"
            
        elif move == move_sets[1]: #down
            nc = (coords[0] + 1, coords[1])
            if prev == "hor":
                cost = cost + 600
            else:
                cost = cost + 100
            next_prev = "ver"
            
        elif move == move_sets[2]: #right
            nc = (coords[0], coords[1] + 1)
            if prev == "ver":
                cost = cost + 600
            else:
                cost = cost + 100
            next_prev = "hor"
            
        elif move == move_sets[3]: #left
            nc = (coords[0], coords[1] - 1)
            if prev == "ver":
                cost = cost + 600
            else:
                cost = cost + 100
            next_prev = "hor"
            
        if nc == (n-1, n-1):
            if answer == 0:
                answer = cost
            else:
                answer = min(answer, cost)
            continue
                
        candi_dests = ((nc[0] - 1, nc[1]), (nc[0] + 1, nc[1]), (nc[0], nc[1] + 1), (nc[0], nc[1] - 1))
        dests = [move_sets[candi_dests.index(item)] for item in candi_dests if (not item in next_visited) and (item[0] in range(n)) and (item[1] in range(n)) and (board[item[0]][item[1]] == 0)]
        
        for item in dests:
            stack.append((nc, item, next_visited, next_prev, cost))
    
    return answer
```

```python
board = [[0,0,0],[0,0,0],[0,0,0]]

solution(board)
```
